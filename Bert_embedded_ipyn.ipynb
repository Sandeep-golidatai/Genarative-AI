```python
token1=['[CLS]']+tokens+['[SEP]']
token1
```

```python
token2=['[CLS]'] + tokens + ['[SEP]']+ ['[CLS]'] + tokens + ['[PAD]']
token2
```

**Attention Mask**

```python
attention_mask=[1     if i!='[PAD]'     else 0           for i in token2]
attention_mask
```

**Convet tokens to ids**

```python
token_ids=tokenizer.convert_tokens_to_ids(token2)
token_ids
```

```python
import torch
ids=torch.tensor(token_ids).unsqueeze(0)
mask=torch.tensor(attention_mask).unsqueeze(0)
```

```python
ids
```

```python
mask
```

```python
out=bert_model(ids,mask)
out
```
