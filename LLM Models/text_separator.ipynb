```python
with open('/content/IndiaUS.txt') as f:
  mytext_file=f.read()
```

```python
from langchain_text_splitters import CharacterTextSplitter


text_splitter=CharacterTextSplitter(
    separator='\n\n',
    chunk_size=1000
)
texts=text_splitter.create_documents([mytext_file])
texts
```

```python
print(texts[0])
```

```python
print(texts[0].page_content)
```

```python
print(len(texts[0].page_content))
```

```python
chunks=[len(i.page_content)        for i in texts]
print(sum(chunks))
```
