```python
from transformers import BertTokenizer,BertForMaskedLM
import torch
import numpy as np
```

```python
bert_tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
bert_mlm=BertForMaskedLM.from_pretrained('bert-base-uncased')
```

```python
print(bert_tokenizer)
```

```python
print(bert_mlm)
```

```python
dir(bert_tokenizer)
```

```python
dir(bert_mlm)
```

```python
masked_sentence='droupadi murmu is [MASK] of India'

spl_token_text=bert_tokenizer.tokenize('[CLS]'+masked_sentence+'[SEP]')
spl_token_text
```

```python
id_for_token=bert_tokenizer.convert_tokens_to_ids(spl_token_text)
id_for_token
```

```python
mask_id=[1]*len(id_for_token)
mask_id
```

```python
ids_tensor=torch.tensor([id_for_token])
mask_tensor=torch.tensor([mask_id])
print(ids_tensor)
print(mask_tensor)
```

```python
outputs=bert_mlm(ids_tensor,mask_tensor)
outputs[0].shape
```

```python
logits=outputs.logits

iden_mask=spl_token_text.index('[MASK]')
masked_token_logits=logits[0,iden_mask,:]
predict_token_id=torch.argmax(masked_token_logits).item()
predict_id_to_token=bert_tokenizer.convert_ids_to_tokens(predict_token_id)
predict_id_to_token
```

```python
unmasked_is=masked_sentence.replace('[MASK]',predict_id_to_token)
unmasked_is
```
